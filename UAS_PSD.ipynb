```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dir(plt)
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

# Membaca File CSV

```python
#1 Membaca File CSV
df = pd.read_csv('https://raw.githubusercontent.com/FiktiTS/PSD/refs/heads/main/PRSA_Data_Dongsi_20130301-20170228.csv')
df.duplicated()
df.describe()
```

## DATA MINING

```python
# Menampilkan informasi umum tentang dataset
print(df.info())

# Menampilkan 5 baris pertama
print(df.head())

# Mengecek jumlah nilai yang hilang
print(df.isnull().sum())

# Menghapus baris dengan nilai yang hilang (jika diperlukan)
df_cleaned = df.dropna()

# Statistik deskriptif dari dataset
print(df_cleaned.describe())

# Menyimpan data yang telah dibersihkan ke file CSV
df_cleaned.to_csv('cleaned_data.csv', index=False)
```

# Visualisasi Histogram

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_cleaned.iloc[:, 1], bins=30, kde=True, color='blue')
plt.xlabel('Feature 1')
plt.ylabel('Frequency')
plt.title('Histogram Feature 1')
plt.show()
```

# Visualisasi Box Plot

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=df_cleaned.iloc[:, 1], color='green')
plt.ylabel('Feature 1')
plt.title('Box Plot Feature 1')
plt.show()
```

# Visualisasi Bar Plot

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=df_cleaned.index[:20], y=df_cleaned.iloc[:20, 1], color='red')
plt.xlabel('Index')
plt.ylabel('Feature 1')
plt.title('Bar Plot Feature 1 (Sample)')
plt.xticks(rotation=45)
plt.show()
```
